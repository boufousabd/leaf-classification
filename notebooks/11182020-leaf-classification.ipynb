```python
import numpy as np
import pandas as pd
```

```python
leaf_train_df = pd.read_csv("../data/raw/train.csv",index_col='id')
leaf_test_df = pd.read_csv("../data/raw/test.csv",index_col='id')
```

```python
leaf_train_df
```

```python
leaf_test_df
```

```python
train_data, train_target = leaf_train_df[leaf_train_df.columns.difference(['species'])], leaf_train_df['species']
```

## 192 dimensions, 99 classes, 990 données train, 594 données de test (1,584 en total)

## SVM Linear

```python
from sklearn import svm
from sklearn.model_selection import GridSearchCV
```

```python
C = [1, 10, 100, 1000]
```

```python
parameters_grid = [
    {'C': [1, 10, 100, 1000], 'kernel': ['linear']},
    {'C': [1, 10, 100, 1000], 'kernel': ['poly'], 'degree':list(range(2,9)), 'coef0':list(np.linspace(0.000001, 2, num=10))},
    {'C': [1, 10, 100, 1000], 'kernel': ['rbf'], 'gamma':['scale','auto'], 'gamma':list(np.linspace(0.000001, 2, num=10))},
    {'C': [1, 10, 100, 1000], 'kernel': ['sigmoid'], 'coef0':list(np.linspace(0.000001, 2, num=10))}
]
```

```python
svc = svm.SVC()
clf = GridSearchCV(svc, parameters_grid, cv=10, n_jobs=-1)
clf.fit(train_data, train_target)
```

```python
best_params = clf.best_params_
best_params
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
```

```python
clf_linear = svm.SVC(kernel= 'linear', C = 100)
scores_linear = cross_val_score(clf_linear, train_data, train_target, cv=10, scoring='f1_macro')
print("F1_Macro Linear",scores_linear.mean()," (+/-) ", scores_linear.std() * 2)
```

```python
clf_rbf = svm.SVC(kernel= 'rbf', gamma = 0.4444452222222222, C = 100)
scores_rbf = cross_val_score(clf_rbf, train_data, train_target, cv=10, scoring='f1_macro')
print("F1_Macro Rbf",scores_rbf.mean()," (+/-) ", scores_rbf.std() * 2)
```

## Gradient Descent

```python
#alpha, optimal,
from sklearn.linear_model import SGDClassifier
tuned_parameters = [{'alpha': list(np.linspace(0.000001, 2, num = 10))}]
clf_sgd = SGDClassifier(max_iter=10000, penalty='l1', shuffle=True, loss='hinge', tol=1e-3)
clf_sgd_grid_search = make_pipeline(StandardScaler(), GridSearchCV(clf_sgd, tuned_parameters, scoring = 'f1_macro', cv=10))
clf_sgd_grid_search.fit(train_data, train_target)
```

```python
best_param = clf_sgd_grid_search.named_steps['gridsearchcv'].best_params_
display(best_param)
```

```python
clf_sgd_ = SGDClassifier(max_iter=10000, penalty='l1', loss='hinge', tol=1e-3, alpha=1e-06)
scores_sgd_ = cross_val_score(clf_sgd_, train_data, train_target, cv=10, scoring='f1_macro')
print("F1_Macro SGD", scores_sgd_.mean()," (+/-) ", scores_sgd_.std() * 2)
```
