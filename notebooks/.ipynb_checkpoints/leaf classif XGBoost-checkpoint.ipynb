```python
from imblearn.under_sampling import RandomUnderSampler
import xgboost as xgb
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
 train_bd = pd.read_csv('train.csv')
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(train_bd.iloc[:,2:195], train_bd['species'], 
                                                train_size=0.5,stratify = train_bd['species'])
```

```python
df_train = pd.concat([xtrain, ytrain], axis=1)
print(len(df_train))
df_train.head()
```

```python
# Check for class imbalance
df_train.species.value_counts()
```

```python
from collections import Counter
rus = RandomUnderSampler(random_state=5)
                    
```

```python
X_rus, y_rus = rus.fit_resample(xtrain, ytrain)
len(Counter(y_rus))
```

```python
Counter(y_rus)
```

```python
from imblearn.over_sampling import SMOTE



smote = SMOTE(random_state=5)

X_resampled, y_resampled = smote.fit_resample(X_rus, y_rus)
```

```python
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
# Build a pipeline for the preprocessing steps
pipe_prepro = Pipeline([('scaler', StandardScaler()),
                        ('pca', PCA(n_components=192))])
```

```python
# Fit the pipeline and transform the training dataset
X_train_trans = pipe_prepro.fit_transform(X_resampled)
```

```python
X_train_trans
```

```python
y_train_res = y_resampled.copy()
```

```python
y_train_res
```

```python
import warnings
warnings.filterwarnings(action='ignore')
# Instantiate the classifier
xg = xgb.XGBClassifier()
```

```python
import warnings
warnings.filterwarnings(action='ignore')
```

```python

# Fit the classifier to the training data
xg.fit(X_train_trans, y_train_res)
```

```python
from sklearn.metrics import (accuracy_score, confusion_matrix, 
                             classification_report, roc_curve, auc)
```

```python
# Define a function to print selected model metrics
def print_metrics(y_true, y_pred):
    '''Print accuracy score, confusion matrix, and classification report.
    
    Keyword arguments:
    y_true: groud-truth labels
    y_pred: predicted labels
    
    Dependencies: 
    sklearn.metrics.accuracy_score
    sklearn.metrics.confusion_matrix
    sklearn.metrics.classification_report
    '''
    acc_score = accuracy_score(y_true, y_pred)
    conf_matrix = confusion_matrix(y_true, y_pred)
    class_report = classification_report(y_true, y_pred)
    print('Accuracy:', acc_score)
    print('-------')
    print('Confusion matrix:')
    print(conf_matrix)
    print('-------')
    print('Classification report:')
    print(class_report)
```

```python
# View metrics for the training data
y_pred = xg.predict(X_train_trans)

print_metrics(y_train_res, y_pred)
```

```python
X_val, X_test, y_val, y_test = train_test_split(xtest, ytest, test_size=0.5,
                                                random_state=1,
                                                stratify=ytest)
```

```python
# Use the pipeline to transform the validation and test datasets
X_val_trans = pipe_prepro.transform(X_val)
# View metrics for validation data
y_pred = xg.predict(X_val_trans)
```

```python
# View full metrics for the validation set
y_val_pred = xg.predict(X_val_trans)
print_metrics(y_val, y_val_pred)
```

```python
X_test_trans = pipe_prepro.transform(xtest)
```

```python
# View full metrics for the validation set
y_pred_test_xgb = xg.predict(X_test_trans)
print_metrics(ytest, y_pred_test_xgb)
```
